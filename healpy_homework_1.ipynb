```python
def radius(x = 0):
  radius = np.radians([x])
  return radius
```

```python
def vec(x,y):
  pos = np.deg2rad([x, 90 - y])
  theta = pos[1]
  phi = pos[0]
  vector = hp.ang2vec(theta, phi)
  return vector
```

## **F)**

```python
#|b| <= 20
def mascara(b1 = 0, b2 = 0):
  b_1 = np.radians(90 + b1) 
  b_2 = np.radians(90 + b2)
  npix = hp.nside2npix(nside_mapa)
  m_mask = hp.ma(mapa)
  mask = np.zeros(npix, dtype = np.bool)
  pix_theta, pix_phi = hp.pix2ang(nside_mapa, np.arange(npix))
  mask[pix_theta > b_1] = 1
  mask[pix_theta < b_2] = 1
  m_mask.mask = np.logical_not(mask)
  return m_mask
```

```python
mask20 = mascara(20, -20)
```

```python
hp.mollview(mask20, coord = 'G', norm = 'hist', min = -1, max = 1)
```

```python
#|b| <= 30
mask30 = mascara(30, -30)
```

```python
hp.mollview(mask30, coord = 'G', norm = 'hist', min = -1, max = 1)
```
